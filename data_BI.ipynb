```python
import matplotlib.pyplot as plt
import seaborn as sns

# Colonnes des montants produits
montants = ['MntWines', 'MntFruits', 'MntMeatProducts', 
            'MntFishProducts', 'MntSweetProducts', 'MntBakedProds']

# ── 1. HISTOGRAMMES ──────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Histogrammes des montants par produit', fontsize=16, fontweight='bold')

for ax, col in zip(axes.flatten(), montants):
    ax.hist(df[col], bins=30, color='steelblue', edgecolor='white')
    ax.set_title(col)
    ax.set_xlabel('Montant (€)')
    ax.set_ylabel('Fréquence')

plt.tight_layout()
plt.show()

# ── 2. BOXPLOTS ───────────────────────────────────────────
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Boxplots des montants par produit', fontsize=16, fontweight='bold')

for ax, col in zip(axes.flatten(), montants):
    ax.boxplot(df[col], patch_artist=True)
    ax.set_title(col)
    ax.set_ylabel('Montant (€)')

plt.tight_layout()
plt.show()

# ── 3. BOXPLOT COMPARATIF (tous ensemble) ─────────────────
plt.figure(figsize=(12, 6))
df[montants].boxplot(patch_artist=True)
plt.title('Comparaison des montants par produit', fontsize=14, fontweight='bold')
plt.ylabel('Montant (€)')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

montants = ['MntWines', 'MntFruits', 'MntMeatProducts', 
            'MntFishProducts', 'MntSweetProducts', 'MntBakedProds']

# Créer le total par client
df['TotalMnt'] = df[montants].sum(axis=1)

campagnes = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
             'AcceptedCmp4', 'AcceptedCmp5']

for cmp in campagnes:
    groupe_refuse  = df[df[cmp] == 0]['TotalMnt']
    groupe_accepte = df[df[cmp] == 1]['TotalMnt']
    
    t_stat, p_value = stats.ttest_ind(groupe_refuse, groupe_accepte)
    
    resultat = "✅ Impact réel" if p_value < 0.05 else "❌ Pas d'impact"
    print(f"{cmp} → p-value: {p_value:.4f} → {resultat}")
```

```python
from scipy.stats import chi2_contingency

campagnes = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
             'AcceptedCmp4', 'AcceptedCmp5']

for cmp in campagnes:
    # Créer le tableau croisé pays vs campagne
    tableau = pd.crosstab(df['Country'], df[cmp])
    
    # Chi² test
    chi2, p_value, dof, expected = chi2_contingency(tableau)
    
    resultat = "✅ Lien réel" if p_value < 0.05 else "❌ Pas de lien"
    print(f"{cmp} → p-value: {p_value:.4f} → {resultat}")
```
